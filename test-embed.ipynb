```python
from langchain.document_loaders import PyPDFLoader
from langchain.text_splitter import CharacterTextSplitter

loader = PyPDFLoader('./files/admin_menual.pdf')

splitter = CharacterTextSplitter(
    # separator="\n"
    chunk_size=600,
    chunk_overlap=100
)

docs = loader.load_and_split(text_splitter=splitter)

print(docs)

```

```python
import tiktoken

docs_text = ""

for doc in docs:
    docs_text += doc.page_content

def num_tokens_from_docs(docs):
    enconding = tiktoken.get_encoding("cl100k_base")
    num_tokens = len(enconding.encode(docs))
    return num_tokens

res = num_tokens_from_docs(docs_text)

print(f"res1 result : {res}")

```
